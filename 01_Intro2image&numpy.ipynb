```python
# Load the image
data = np.array(plt.imread('cat.jpg'))

# Set the red channel in this part of the image to 0
data[0:2000, 0:2000, 0] = 0

# Set the green channel in this part of the image to 255
data[0:2000, 0:2000, 1] = 255

# Set the blue channel in this part of the image to 0
data[0:2000, 0:2000, 2] = 0

# Visualize the result
plt.imshow(data)
plt.show()
```

### Grayscale image


```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```

```python
img = plt.imread('cat.jpg')
cat = rgb2gray(img)
```

```python
cat.shape
```

```python
plt.imshow(cat, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
cat[250, 200]
```

```python
cat[2000:3000, 3000:4000] = 200
plt.imshow(cat, cmap = plt.get_cmap('gray'))
plt.show()
```
