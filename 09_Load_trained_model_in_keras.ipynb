# Using a pre trained model

## ` Importing libraries`

```python
from tensorflow import keras
from keras.models import load_model
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
```

## ` Load the model`

```python
model = load_model('/content/cats_and_dogs_small_1.h5')
```

##`Download and process a image for prediction`

```python
!wget https://images.theconversation.com/files/350865/original/file-20200803-24-50u91u.jpg
!wget https://images.theconversation.com/files/319375/original/file-20200309-118956-1cqvm6j.jpg
```

```python
cat_img = image.load_img('cat.jpg' , target_size=(150,150))
dog_img = image.load_img('dog.jpg' , target_size=(150,150))
plt.imshow(dog_img)
```

```python
plt.imshow(cat_img)
```

```python
# Chaning images to numpy array
x_cat = image.img_to_array(cat_img)
x_dog = image.img_to_array(dog_img)

# Normalizing channel values
x_cat /= 255.
x_dog /= 255.

# Reshaping arrays so model can predict them
# Method 1 :
x_cat = np.reshape(x_cat , (1,) + x_cat.shape)
x_dog = np.reshape(x_dog , (1,) + x_dog.shape)

# Mehod 2 :
# x_cat = np.expand_dims(x_cat , axis=0)
# x_dog = np.expand_dims(x_dog , axis=0)
```

## ` Predicting`
No idead how to actually interpret the result

```python
cat_pred_proba = model.predict(x_cat)
cat_pred = np.argmax(cat_pred_proba)
dog_pred_proba = model.predict(x_dog)
dog_pred = np.argmax(dog_pred_proba)

print(f'Cat prediction probablites : {cat_pred_proba}\n Cat prediction : {cat_pred}')
print(f'Dog prediction probablites : {dog_pred_proba}\n Dog prediction : {dog_pred}')
```
